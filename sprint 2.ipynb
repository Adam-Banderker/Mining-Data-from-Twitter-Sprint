```python
tweets = tw.Cursor(api.search, 
                           q=new_search,
                           lang="en",
                           since=date_since).items(50) 

users_locs = [[tweet.user.created_at,tweet.user.screen_name, tweet.user.statuses_count] for tweet in tweets]
users_locs
```

```python
#Created a Pandas Dataframe From A List of Tweet Data
tweet_text = pd.DataFrame(data=users_locs, 
                    columns=["Datetime","User", "Tweets count"])
tweet_text

```

```python
tweet_text.to_csv("Sprint2.csv")
```
